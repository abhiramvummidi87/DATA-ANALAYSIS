```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '3',label ='CompanyA',marker='.',ms='20',mec='k')
#plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label ='CompanyB',marker = 'H')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("iris")
print(tips)
sns.scatterplot(x="sepal_length",y="sepal_width",data=tips,label='iris')
plt.legend(loc='best')
plt.title("Scatter plot of Total bill vs tip")
plt.xlabel("sepal_length ($)")
plt.ylabel("sepal_width($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips=sns.load_dataset("tips")
print(tips)

sns.barplot(x="day",y="total_bill",data=tips,label="Hotel")
plt.title("Scatter plot of Total Bill vs, Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(loc='best')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plot
iris=sns.load_dataset("iris")
correlation_matrix=iris.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("correlation heatmap of iris dataset")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("iris")
print(tips)
sns.countplot(x="sepal_length",data=tips,label='iris')
plt.legend(loc='best')
plt.title("Scatter plot of Total bill vs tip")
plt.xlabel("sepal_length ($)")
plt.ylabel("sepal_width($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
print(tips)
sns.lmplot(x="total_bill", y="tip",data=tips)
plt.title("linear regression of total bill and tip")
plt.xlabel("total_bill ($)")
plt.ylabel("tip($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
print(titanic)
sns.countplot(x="class",data=titanic)
g=sns.FacetGrid(titanic,col="class")
g.map(sns.histplot,"age")
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Age Distribution by Passenger Class")
plt.show()
```
